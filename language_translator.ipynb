language translation by entering language want to translate 
It will automatically detect your enter language from input.

```python
from textblob import TextBlob
from langdetect import detect # !pip install langdetect
import googletrans  #!pip install googletrans
import pandas as pd
data=(googletrans.LANGUAGES)
lang=str(input("enter the language to convert(EX-english) "))
lang_low=lang.lower()
key = next(key for key, value in data.items() if value == lang_low)
to_lang=str(key)
for i in c:
  a=detect(i)
  word=TextBlob(i)
  z=word.translate(from_lang=a, to=to_lang)
  print(z)
```
